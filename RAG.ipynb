```python
import pandas as pd
```

# 1. Read the file

- 데이터 엔티티를 추출한 파일을 읽는다

- column의 이름을 다시 한번 지정함

```python
df = pd.read_excel("mouse_disease_model_db.xlsx")
df.columns = ['No', 'Title', 'Abstract', 'Disease', 'Model', 'Species', 'Key Target', 'Keywords']
```

2. Create function

    - 검색결과를 나타내줄 함수를 만든다
   
    - Disease와 Key Target에서 함수에 입력된 단어를 찾는다

```python
def biological_rag_system(query):
    results = df[df['Disease'].str.strip().str.contains(query, case=False, na=False) |
                 df['Key Target'].str.strip().str.contains(query, case=False, na=False)]
    if results.empty:
        return "해당 질병에 대한 적합한 모델 데이터를 찾을 수 없습니다"
    top_match = results.iloc[0]
    response = f"""
    [검색 결과 분석]
    - 추천 모델: {top_match['Model']} ({top_match['Species']}
    - 핵심 기전: {top_match['Abstract']}
    - 주요 타겟: {top_match['Key Target']}
    - 근거 논문: {top_match['Title']}
    """
    return response
```

```python
print(biological_rag_system("Glioblastoma"))
```

```python
print(biological_rag_system("liver"))
```

```python
print(biological_rag_system('Parkinson'))
```
